```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('punkt_tab')
input = "I am learning Natural Language Processing. How about you?"
output1 = word_tokenize(input)
print(output1)
```

```python
input = "I am learning Natural Language Processing. How about you?"
output2 = sent_tokenize(input)
print(output2)
```

```python
text = """Hi My name is Thirumurugan B. studying in alliance university. currently in sixth semester.
Hi My name is Thirumurugan B. studying in alliance university. currently in sixth semester."""

output1 = word_tokenize(text)
print(output1)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "I am learning Natural Language Processing. How about you?"
doc = nlp(input)
for token in doc:
  print(token.text)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp("This is a red ball.")
print([(w.text, w.pos_) for w in doc])
```
