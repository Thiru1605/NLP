```python
import re

# Sample text for demonstration
raw_text = """Hello there! Check out my website: https://example.com and email me at test@email.com. This is bold text with some numbers like 123 and 45.67. Another URL: www.anothersite.org. Don't forget my other email: user.name@sub.domain.co.uk! And some HTML: Paragraph text. Price is $100.00!"""

print("Original Text \n:")
print(raw_text)
print("\n---")
```

```python
# Regex to remove email addresses
cleaned_text_no_emails = re.sub(r'\S*@\S*\s?', '', raw_text)

print("Text after removing emails:")
print(cleaned_text_no_emails)
print("\n---")
```

```python

# Regex to remove URLs
# This pattern covers http/https, www. and direct domain names.
cleaned_text_no_urls = re.sub(r'http\S+|www\S+|\S+\.(com|org|net|gov|edu|uk|co\.uk)\b', '', cleaned_text_no_emails)

print("Text after removing URLs:")
print(cleaned_text_no_urls)
print("\n---")
```

```python
# Regex to remove HTML tags
cleaned_text_no_html = re.sub(r'<.*?>', '', cleaned_text_no_urls)

print("Text after removing HTML tags:")
print(cleaned_text_no_html)
print("\n---")
```

```python

# Regex to remove numbers (integers and decimals)
cleaned_text_no_numbers = re.sub(r'\b\d+\.\d+\b|\b\d+\b', '', cleaned_text_no_html)

print("Text after removing numbers:")
print(cleaned_text_no_numbers)
print("\n---")
```

```python

import string

# Regex to remove punctuation
cleaned_text_no_punc = cleaned_text_no_numbers.translate(str.maketrans('', '', string.punctuation))

print("Text after removing punctuation:")
print(cleaned_text_no_punc)
print("\n---")
```

```python

# Convert all text to lowercase
final_cleaned_text = cleaned_text_no_punc.lower()

print("Final cleaned text (lowercase):")
print(final_cleaned_text)
print("\n---")
```

```python

import re
import string

def clean_text(text):
    # 1. Remove Emails
    text = re.sub(r'\S*@\S*\s?', '', text)
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|\S+\.(com|org|net|gov|edu|uk|co\.uk)\b', '', text)
    # 3. Remove HTML Tags
    text = re.sub(r'<.*?>', '', text)
    # 4. Remove Numbers
    text = re.sub(r'\b\d+\.\d+\b|\b\d+\b', '', text)
    # 5. Remove Punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # 6. Convert to Lowercase
    text = text.lower()
    # 7. Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Test the function with the original raw text
cleaned_full_text = clean_text(raw_text)

print("Original Text:\n", raw_text)
print("\nCleaned Text (using function):\n", cleaned_full_text)
```
